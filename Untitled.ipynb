```python
def MSE(actual, prediction):
    return sum((actual-prediction)**2)/len(prediction)
```

```python
f_initial = f'pts~{string}'
model_initial = ols(formula=f_initial, data=df_stars_workable).fit()
model_initial.summary()
stars_prediction = model_initial.predict(df_stars_workable)
MSE(df_stars_workable['pts'],stars_prediction)
model_initial.summary()
```

```python
#randomizes order of players
df_stars_workable.columns
x = df_stars_workable[['age', 'player_height', 'player_weight', 'country',
       'gp', 'reb', 'ast', 'net_rating', 'oreb_pct', 'dreb_pct',
       'ts_pct', 'bmi', 'vol_shooting']]

df_stars_workable = df_stars_workable.sample(frac=1).reset_index(drop=True)
df_stars_workable
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(x, df_stars_workable['pts'], test_size = 0.2, random_state = 3)



lin = LinearRegression()
lin.fit(X_train, Y_train)
```

```python

y_h_lin_train = lin.predict(X_train)
y_h_lin_test = lin.predict(X_test)
```

```python

print('Train Error Unpenalized Linear Model', np.sum((Y_train - y_h_lin_train)**2)/int(len(Y_train)))
print('Test Error Unpenalized Linear Model', np.sum((Y_test - y_h_lin_test)**2)/int(len(Y_test)))
lin.coef_
coef = dict(zip(list(x.columns),lin.coef_))
print('\n')
print('Coefficients', coef)
print('\n')
print('intercept', lin.intercept_)
print('R_squared', lin.score(X_test, Y_test))
```
